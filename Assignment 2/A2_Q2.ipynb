```python
#the path from where we have to access our data.
file_path='gdrive/MyDrive/MLA2_Q2_dataset'
```

```python
#gives the list of names of images present in the folder.

files = os.listdir(file_path)                       
print(files)
```

```python
#function to select random Images

def random_image(files):
  random_list=random.sample(files, 3)               #selects random three images from list "files"
  return random_list
```

##Plotting Original Images

```python
random_list=random_image(files)
print("Images selected : {}".format(random_list))

plt.figure(figsize=(20,20))
for i in range(len(random_list)):
  imgpath= os.path.join(file_path,random_list[i])
  # print(imgpath)
  image = cv.imread(imgpath)
  plt.subplot(1, 3, i+1)
  plt.imshow(image)
```

##Resizing the Images

```python
random_list=random_image(files)
print("Images selected : {}".format(random_list))

plt.figure(figsize=(20,20))
for i in range(len(random_list)):
  imgpath= os.path.join(file_path,random_list[i])
  # print(imgpath)
  image = cv.imread(imgpath)
  resize_image = cv.resize(image, (90,90))
  plt.subplot(1, 3, i+1)
  plt.imshow(resize_image)
```

##b) Padding

```python
random_list=random_image(files)
print("Images selected : {}".format(random_list))

plt.figure(figsize=(20,20))
for i in range(len(random_list)):
  imgpath= os.path.join(file_path,random_list[i])
  # print(imgpath)
  image = cv.imread(imgpath)
  resize_image = cv.copyMakeBorder(image, 60, 60, 60, 60, cv.BORDER_CONSTANT, None, value = 0)
  plt.subplot(1, 3, i+1)
  plt.imshow(resize_image)

# .copyMakeBorder()
# image = cv.copyMakeBorder(image, 10, 10, 10, 10, cv2.BORDER_CONSTANT, None, value = 0)
```

#d) GrayScale

#e) Contrast

#####For contrast of image I have used addWeighted Function of OpenCV Library.
#####There is no direct funtion in openCV by we can directly change the contrast rather there is a equation present which is  "new_image=  alpha * old_image + beta"
#####Syntax: addWeighted(src1, alpha, src2, beta, gamma)
#####This syntax will blend 2 images so if we want to blend only single image then we can pass second image source as zeros using numpy and this is what we have done. Since there is no other image, the np.zeros will be used to create an array of the same shape and data type as the original image but the array will be filled with zeros.

```python
random_list=random_image(files)
print("Images selected : {}".format(random_list))

alpha=1.2
beta=20

plt.figure(figsize=(20,20))
for i in range(len(random_list)):
  imgpath= os.path.join(file_path,random_list[i])
  # print(imgpath)
  image = cv.imread(imgpath)
  image_bc = cv.addWeighted(image,alpha,np.zeros(image.shape,image.dtype),0,beta)
  plt.subplot(1, 3, i+1)
  plt.imshow(image_bc)
```

#f)Saturation

###### H denotes hue, S denotes Saturation and V denotes value.
######Value: This is a measure of the brightness of a colour. 
######Saturation: Saturation gives the purity of a colour. A pure colour has no gray mixed in it. Greater the amount of gray mixed in a colour, lesser the saturation. The saturation value is usually measured from 0 to 100% but in OpenCV the scale for saturation is from 0 to 255.
######To summarize, hue is the purest form of the primary as well as the secondary colours. Generally, hue is measured in 360 degrees of a colour circle, but in OpenCV the hue scale is only 180 degrees.

```python
random_list=random_image(files)
print("Images selected : {}".format(random_list))

plt.figure(figsize=(20,20))
for i in range(len(random_list)):
  imgpath= os.path.join(file_path,random_list[i])
  # print(imgpath)
  image = cv.imread(imgpath)
  hsv = cv.cvtColor(image, cv.COLOR_BGR2HSV)
  plt.subplot(1, 3, i+1)
  plt.imshow(hsv)
```
